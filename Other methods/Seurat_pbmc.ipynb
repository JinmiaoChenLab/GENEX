```r
# read the raw counts data
raw_counts <- read.csv("count_matrix_pbmc_full.csv", header= TRUE)

# read the meta data
cell_meta <- read.csv("metadata_cell_pbmc_full.csv", header= TRUE)
gene_meta <- read.csv("metadata_gene_pbmc_full.csv", header= TRUE)
```

```r
# create seurat object
# mydata <- CreateSeuratObject(counts = raw_counts, min.cells = 3, min.genes = 200, project = "mydata_scRNAseq")
mydata <- CreateSeuratObject(counts = raw_counts, project = "mydata_scRNAseq")
```

```r
mydata
```

```r
# adding cell meta data such as batch and cell type
mydata$batch <- cell_meta$batch
mydata$Method <- cell_meta$Method
mydata$CellType <- cell_meta$CellType
```

```r
# split the dataset into a list of batch object
data.list <- SplitObject(mydata, split.by = "batch")

# normalize and identify variable features for each dataset independently
data.list <- lapply(X = data.list, FUN = function(x) {
#     x <- NormalizeData(x)
    x <- FindVariableFeatures(x, nfeatures = 10000)
})

# select features that are repeatedly variable across datasets for integration
features <- SelectIntegrationFeatures(object.list = data.list)
```

```r
immune.anchors <- FindIntegrationAnchors(object.list = data.list, anchor.features = features)
```

```r
all_features <- lapply(data.list, row.names) %>% Reduce(intersect, .) 
immune.combined <- IntegrateData( 
  anchorset = immune.anchors, 
  features.to.integrate = all_features,
  verbose = T
)
```
